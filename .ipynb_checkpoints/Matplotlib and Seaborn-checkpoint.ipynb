```python
#Example Questions:
#1.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
import seaborn as sns
```

```python
style.use("ggplot")
```

```python
values = np.random.rand(10)
plt.figure(figsize = (8,3))
plt.plot(values,'g', label = "Random data", linewidth = 2, linestyle = "--")
plt.title("Random data line chart")
plt.xlabel("X")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2.
x = list(range(0,10))
y = np.random.rand(10)
plt.figure(figsize = (4,4))
plt.scatter(x,y, linewidths = 4)
plt.title("Scatter plot of random data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Practice Questions:
#1.
df = pd.read_csv("mtcars.csv")
plt.figure(figsize = (8,4))
plt.plot(df['mpg'],label = "Mpg")
plt.title("Line chart of Mpg over indexs in the data set")
plt.xlabel("Indexes of dataset")
plt.ylabel("Mpg of the car")
plt.legend()
plt.show()
```

```python
#2.
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (6,3))
plt.hist(df['disp'],label = "Displacement", bins = 15, color = 'green')
plt.title("Distribution of displacement")
plt.xlabel("Displacement")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#3.
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (4,4))
sns.scatterplot(data = df, x ='cyl', y ='carb')
plt.title("Scatter plot cylinder vs carb")
plt.xlabel("Cylinder")
plt.ylabel("Carb")
plt.show()
```

```python
#4.
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (4,4))
for i in df.iloc[:,1:11].columns:
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (7,4))
sns.barplot(data = df, x ='cyl', y ='disp')
plt.title("Bar plot cylinder vs displacement")
plt.xlabel("Cylinder")
plt.ylabel("Displacement")
plt.show()
```

```python
#6.
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (6,5))
sns.violinplot(data = df, x ='cyl', y ='disp')
plt.title("Violenplot cylinder vs displacement")
plt.xlabel("Cylinder")
plt.ylabel("Displacement")
plt.show()
```

```python
#7.
df = pd.read_csv("mtcars.csv")
df.head()
plt.figure(figsize = (6,6))
sns.pairplot(data = df, x_vars = 'disp', y_vars ='hp', kind = 'reg')
plt.title("Violenplot cylinder vs displacement")
plt.xlabel("Cylinder")
plt.ylabel("Displacement")
plt.show()
```
